### Pull Request 이후 추가 수정 및 재요청의 경우

- 동일 브랜치에서 커밋 추가 시 기존 Pull Request의 자동 반영됨. 새로운 Pull Request 생성 불필요함.

### Upstream 저장소의 변경 발생 시
- 로컬 저장소에서 upstream 브랜치의 변경 사항을 fetch 및 merge 또는 rebase하여 동기화 필요함.

### 같은 라인의 코드 동시 수정 상황
- 병합 충돌 발생 가능성 있음. 수동 충돌 해결 후 Pull Request 제출 필요함.

### 대상 저장소의 지속적인 업데이트 반영 필요 시

- 작업 브랜치를 정기적으로 upstream의 최신 버전으로 rebase 또는 merge하여 기반 코드 최신화 필요함.

### "git pull"은 위 상황에 적절하지 않은가?

- 왜냐하면, 이 명령은 “origin(자신의 포크 저장소)”의 main 브랜치를 가져오는 것이지, “upstream(원본 저장소)”의 main 브랜치를 가져오는 것이 아니기 때문.
